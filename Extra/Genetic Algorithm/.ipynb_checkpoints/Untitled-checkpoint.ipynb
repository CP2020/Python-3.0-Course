```python
import random
```

```python
n = 4
p = 100

current_generation = []
new_generation = []
```

```python
def randomP(n,p):
    generation_list = []
    for i in range(p):
        generation_list.append([random.randint(0,n-1) for x in range(n)])
        generation_list[i].append(0)
    return generation_list
```

```python
def fitness(generation_list):
    i = 0
    while i<len(generation_list):
        j = 0
        conflict=0
        while j < len(generation_list[i]):
            k = j+1
            while k<len(generation_list[i]):
                if generation_list[i][j] == generation_list[i][k]:
                    conflict+=1
                if abs(j-k)==abs(generation_list[i][j]-generation_list[i][k]):
                    conflict+=1
                k+=1
            j+=1
        generation_list[i][n]=conflict
        i+=1
    i = 0
    while i<len(generation_list):
        j = i
        while j<len(generation_list):
            if generation_list[i][n] > generation_list[j][n]:
                temp = generation_list[i]
                generation_list[i] = generation_list[j]
                generation_list[j] = temp
            j+=1
        i+=1
    return generation_list
```

```python
def cross_over(generation_list):
    for i in range(0,len(generation_list),2):
        z = 0
        new_kid1 = []
        new_kid2 = []
        while z<n:
            if(z<n//2):
                new_kid1.append(generation_list[i][z])
                new_kid2.append(generation_list[i+1][z])
            else:
                new_kid1.append(generation_list[i+1][z])
                new_kid2.append(generation_list[i][z])
            z+=1
        new_kid1.append(0)
        new_kid2.append(0)
        generation_list.append(new_kid1)
        generation_list.append(new_kid2)
    return generation_list
```

```python
def mutation(generation_list):
    muted_list=[]
    i = 0
    while i<p//2:
        new_rand = random.randint(p//2,p-1)
        if new_rand not in muted_list:
            muted_list.append(new_rand)
            generation_list[new_rand][random.randint(0,n-1)]=random.randint(0,n-1)
            #print("Muted:",new_rand)
            i+=1
    return generation_list
```

```python
current_generation = randomP(n,p)
epoch = 1
while True:
    print("# Epoch",epoch)
    current_generation = fitness(current_generation)
    new_generation = current_generation[0:(p//2)]
    new_generation = cross_over(new_generation)
    new_generation = mutation(new_generation)
    new_generation = fitness(new_generation)
    if new_generation[0][n]==0:
        print("Solution Found:",new_generation[0])
        break
    else:
        #print("Solution Not Found")
        print("Best Solution",new_generation[0])
    current_generation = new_generation
    epoch+=1
```
