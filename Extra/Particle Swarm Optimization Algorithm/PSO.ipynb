```python
import random
import numpy as np
```

```python
"""num_particles = int(input("Please Enter the Number of Particles: "))
num_epoches = int(input("Please Enter the Number of Epoches: "))
low_value= int(input("Please Enter the low value: "))
high_value = int(input("Please Enter the high value: "))"""

num_particles=20
num_epoches=50
low_value=-10
high_value=20

w= 0.5
c1 = 0.5
c2 = 0.9
target = 1
```

```python
def fitness(x):
    return abs((x-2)%2)+1
```

```python
def all_fitness(p_list):
    for i in range(num_particles):
        p_list[i][3] = fitness(p_list[i][0])
    return p_list
```

```python
def population_creator(numP):
    particle_list = []
    for i in range(numP):
        x = random.randint(low_value,high_value)
        s = np.random.random()
        particle_list.append([x,s,x,fitness(x)])
    return particle_list
```

```python
def pbest(p_list):
    for i in range(num_particles):
        if fitness(p_list[i][0])<p_list[i][2]:
            p_list[i][2]=p_list[i][0]
    return p_list
```

```python
def g_best(p_list):
    for j in range(num_particles):
        for i in range(j,num_particles):
            if p_list[j][3]>p_list[i][3]:
                temp = p_list[j]
                p_list[j]=p_list[i]
                p_list[i]=temp
    return p_list
```

```python
def update_positios(p_list):
    for i in range(1,num_particles):
        velocity = (
        ( w * p_list[i][1] ) +
        ((c1 * random.randint(0,1))*(p_list[i][2]-p_list[i][0])) +
        ((c2 * random.randint(0,1))*(p_list[0][0]-p_list[i][0]))
        )
        if fitness(p_list[i][2]>fitness(p_list[i][0]+velocity)):
            if p_list[i][0]+velocity < high_value and p_list[i][0]+velocity > low_value:
                p_list[i][0]=p_list[i][0]+velocity
        else:
            p_list[i][2]=0
    return p_list
```

```python
particles = population_creator(num_particles)
particles = g_best(particles)
for i in particles:
        print(i)
```

```python
for zz in range(5):
    particles = update_positios(particles)
    particles = all_fitness(particles)
    particles = pbest(particles)
    particles = g_best(particles)
    for i in particles:
        print(i)
    print("---------------------------------------------------------------------------------------")
```
