```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,9,7]

plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Simple Graph")
plt.show()
```

```python
x = [1,2,3]
y = [5,9,7]

plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Simple Graph\n Check it out")
plt.show()
```

```python
x = [1,2,3]
y = [5,9,7]

x2 = [1,2,3]
y2= [7,11,9]

plt.plot(x,y)
plt.plot(x2,y2)
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Simple Graph\n Check it out")
plt.show()
```

```python
x = [1,2,3]
y = [5,9,7]

x2 = [1,2,3]
y2= [7,11,9]

plt.plot(x,y, label="first-line")
plt.plot(x2,y2, label="second-line")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Simple Graph\n Check it out")
plt.legend()
plt.show()
```
